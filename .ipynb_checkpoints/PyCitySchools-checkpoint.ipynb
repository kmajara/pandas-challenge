# PyCitySchools Analysis

PyCitySchoools data set consists of data representing close to 40,000 students across 15 schools. On an aggregate level, the average math score is 78.9 and the average reading score is  81.8. On close examination of the data, there are multiple nuances. 

At a school level, charter schools have higher reading and math passing rates, as well as overall passing rates than their district counterparts. 
Overall, there is no significant variability in reading and math scores across the grades . 
In terms of spending, it appears that the average math and reading scores, as well as the overall passing rates are inversely proportional  to the amount of dollars spent per student. 
In terms of school size, small and medium sized schools tend to have higher math, reading, and overall passing rates. While not part of the data sample, a lower teach to student ratio typically observed in smaller schools might be a contributing  factor. In conclusion, small to medium sized charter schools with relatively small budgets perform better in math and reading and have higher overall passing rates. 


```python
#Dependencies and Setup
import pandas as pd
from pathlib import Path

#Files to Load
school_data_source_path = Path("Resources/schools_complete.csv")
student_data_source_path = Path("Resources/students_complete.csv")

#Read School and Student Data File and store into Pandas Dataframe
school_data = pd.read_csv(school_data_source_path)
student_data = pd.read_csv(student_data_source_path)



#Compbine the data into a single dataset
school_data_complete =pd.merge(student_data, school_data, how = "left", on = ["school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools

school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
#Calculate the total number of students

student_count = school_data_complete["student_name"].size
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
# Calculate the average (mean) math score

average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
#calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# percentage of students who passed math 

passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count/float(student_count) * 100
passing_math_percentage
```

```python
# Percentage of students who passed reading 

passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/float(student_count) * 100
passing_reading_percentage
```

```python
# Percentage of students that passed math AND reading 

passing_math_reading_count = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"]>=70)].count()["student_name"]
district_overall_passing_rate = passing_math_reading_count/float(student_count) * 100
district_overall_passing_rate
```

```python
#High Level snapshot of the district's key metrics in a DataFrame 

district_summary = pd.DataFrame({
    "Total Schools":[school_count],
    "Total Students":[student_count],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_reading_score],
    "% Passing Math":[passing_math_percentage],
    "% Passing Reading":[passing_reading_percentage],
    "Overall Passing":[district_overall_passing_rate]
})


# Formatting [formatting all of the summary items for a cleaner look]

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary["Overall Passing"] = district_summary["Overall Passing"].map("{:,.2f}".format)

district_summary
```

## School Summary

```python
# Selecting the type of school from school_data

school_types = school_data.set_index(["school_name"])["type"]
print(school_types)
```

```python
# Calculate the total student count per school from school data

per_school_counts = school_data.set_index(["school_name"])["size"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school from school_data
per_school_budget = school_data.set_index(["school_name"])["budget"]
per_school_capita = per_school_budget/per_school_counts
per_school_capita.head(15)
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()

print("Average Math Score per School:")
print(per_school_math)

print("\nAverage Reading Score per School:")
print(per_school_reading)
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
students_passing_math = school_data_complete[school_data_complete['math_score'] >= 70]
school_students_passing_math = students_passing_math.groupby('school_name').size()

print("Number of Students Passing Math per School:")
print(school_students_passing_math)
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete
students_passing_reading = school_data_complete[school_data_complete["reading_score"]>=70]
school_students_passing_reading = students_passing_reading.groupby('school_name').size()

print("Number of students passing reading per school")
print(school_students_passing_reading)
```

```python
# calculating the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()

print("Number of students passing both math and reading by school")
print(school_students_passing_math_and_reading)
```

```python
# calculating the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
school_overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100

print("Per School Passing rate - math\n")
print(per_school_passing_math)
print("\n Per School Passing rate - reading\n")
print(per_school_passing_reading)
print("\noverall passing rate\n")
print(school_overall_passing_rate)
```

```python
# Creating a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School type": school_types,
    "Total students": per_school_counts,
    "Total school budget": per_school_budget,
    "Per student budget": per_school_capita,
    "Number passing math": school_students_passing_math,
    "Average math score": per_school_math,
    "% passing math": per_school_passing_math,
    "Number passing reading": school_students_passing_reading,
    "Average reading score": per_school_reading,
    "% passing reading": per_school_passing_reading,
    "% overall passing": school_overall_passing_rate})


#Formatting
per_school_summary["Total students"] = per_school_summary["Total students"].map("{:,.0f}".format)
per_school_summary["Total school budget"] = per_school_summary["Total school budget"].map("${:,.2f}".format)
per_school_summary["Per student budget"] = per_school_summary["Per student budget"].map("${:,.2f}".format)
per_school_summary["Number passing math"] = per_school_summary["Number passing math"].map("{:,.0f}".format)
per_school_summary["Number passing reading"] = per_school_summary["Number passing reading"].map("{:,.0f}".format)
per_school_summary["Average math score"] = per_school_summary["Average math score"].map("{:.2f}".format)
per_school_summary["Average reading score"] = per_school_summary["Average reading score"].map("{:.2f}".format)
per_school_summary["% passing math"] = per_school_summary["% passing math"].map("{:.2f}".format)
per_school_summary["% passing reading"] = per_school_summary["% passing reading"].map("{:.2f}".format)
per_school_summary["% overall passing"] = per_school_summary["% overall passing"].map("{:.2f}".format)


# Display the DataFrame
per_school_summary
```

## Lowest Perforiming Schools - in ascending order (by % overall passing)

```python
bottom_schools = per_school_summary.sort_values("% overall passing")
bottom_schools
```

## Math Scores by Grade

```python
#collect required fields
grade_scores_group = school_data_complete.groupby(["school_name", "grade"])

#calculate the average
average_math_by_grade = grade_scores_group["math_score"].mean()

#create a data frame showing the average math scores by grade
average_math_by_grade_df = pd.DataFrame(average_math_by_grade)

#reshape the data presentation
grade_order = ['9th', '10th', '11th', '12th'] 
pivoted_math_by_grade_df = average_math_by_grade_df.pivot_table(values='math_score', index='school_name', columns='grade',aggfunc='first')[grade_order]
pivoted_math_by_grade_df.head(10)
```

## Reading Scores by Grade

```python

#calculate the average using the previously created variable [grade_scores_group]
average_reading_by_grade = grade_scores_group["reading_score"].mean()

#create a data frame showing the average reading scores by grade
average_reading_by_grade_df = pd.DataFrame(average_reading_by_grade)

#reshape the data presentation
pivoted_reading_by_grade_df = average_math_by_grade_df.pivot_table(values='math_score', index='school_name', columns='grade',aggfunc='first')[grade_order]
pivoted_reading_by_grade_df.head(10)
```

## Scores by School Spending

```python
#Convert required columns' data type to float
import re

columns_to_convert = ["Total students","Total school budget","Per student budget","Number passing math","Number passing reading","Average math score","Average reading score", "% passing math", "% passing reading","% overall passing"]

for col in columns_to_convert:
    per_school_summary[col] = per_school_summary[col].str.replace(',', "", regex=True)
    per_school_summary[col] = per_school_summary[col].str.replace(r'\$', '', regex=True)
    per_school_summary[col] = per_school_summary[col].astype(float)

school_spending_df = per_school_summary
```

```python
#create spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

#use pd.cut to categorize spending ranges
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per student budget"], spending_bins, labels=labels)

school_spending_df.head()
#school_spending_df['Set_of_Numbers'] = df['Set_of_Numbers'].fillna(0)
```

```python
#calculate means scores per spending range

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average math score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average reading score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passing math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passing reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% overall passing"].mean()

#creating a df
spending_summary = pd.DataFrame({
    "Average math score": spending_math_scores,
    "Average reading score":spending_reading_scores,
    "% passing math": spending_passing_math,
    "% passing reading": spending_passing_reading,
    "% overall passing": overall_passing_spending})

spending_summary.head()
```

## Scores by School Size

```python
#Establish bins

size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# assign the categories and bins 

per_school_summary["School Size"] = pd.cut(per_school_summary["Total students"], bins = size_bins, labels = size_labels)
per_school_summary.head()
```

```python
# Calculate averages for needed columns.
math_scores_size = per_school_summary.groupby(["School Size"], observed=False)["Average math score"].mean()
reading_scores_size = per_school_summary.groupby(["School Size"], observed=False)["Average reading score"].mean()
passing_math_size = per_school_summary.groupby(["School Size"], observed=False)["% passing math"].mean()
passing_reading_size = per_school_summary.groupby(["School Size"], observed=False)["% passing reading"].mean()
overall_passing_size = per_school_summary.groupby(["School Size"], observed=False)["% overall passing"].mean()
```

```python
#creating the size_summary dataframe
size_summary = pd.DataFrame({
    "Average math score": math_scores_size,
    "Average reading score": reading_scores_size,
    "% passing math": passing_math_size,
    "% passing reading": passing_reading_size,
    "% overall passing": overall_passing_size
})

size_summary
```

```python
# calculate averages for needed columns
avg_math_score_by_type = per_school_summary.groupby(["School type"])["Average math score"].mean()
avg_reading_score_by_type = per_school_summary.groupby(["School type"])["Average reading score"].mean()
avg_percent_passing_math_by_type = per_school_summary.groupby(["School type"])["% passing math"].mean()
avg_percent_passing_reading_by_type = per_school_summary.groupby(["School type"])["% passing reading"].mean()
avg_percent_overall_passing_by_type = per_school_summary.groupby(["School type"])["% overall passing"].mean()
```

## Scores by School Type

```python
#creating the type_summary Dataframe

type_summary = pd.DataFrame({
    "Average math score": avg_math_score_by_type,
    "Average reading score": avg_reading_score_by_type,
    "% passing math": avg_percent_passing_math_by_type,
    "% passing reading": avg_percent_passing_reading_by_type,
    "% overall passing": avg_percent_overall_passing_by_type
})

# Display results
type_summary.head()
```
